Среднее значение, медиана:

```python
def mean(array):
    return Decimal(sum(array)/len(array))

def mediana(array):
    new_array = sorted(array)
    length = len(new_array)
    if (length % 2 == 0):
        return (new_array[length//2] + new_array[length//2 - 1]) / 2
    return new_array[(length-1)//2]
```

Дисперсия, Стандартное отклонение:

```python
def sample_dispertion(array):
    ave = mean(array)
    summa = Decimal(0.0)
    for cell in array:
        summa += ((Decimal(cell) - ave)**2)
    return Decimal(summa/(len(array)-1))

def sigma(array):
    return sample_dispertion(array).sqrt()
```

Квантили уровней 0.01, 0.05, 0.95, 0.99:

```python
def quantil(array, proc):
    new_array = sorted(array)
    return new_array[int(np.floor((len(new_array)-1) * proc))]
```

Выявление экстремальных наблюдений:

Выявление экстремальных наблюдений, не имеющих содержательной экономической интерпретации, рассматривается на примере работы с данными столбца "k7", соответсвующим значениям
качества кредиторской задолженности предприятий. Экстремальные наблюдения выявляются с помощью теста Хампеля (в качестве множителя в формуле рассчета интервала неэкстремальных наблюдений
берется число 6.4 таким образом, чтобы общее число исключенных из выборки после выявления экстремальных данных не превышало 8% от общего числа наблюдений).

```python
def HampelTest(col):
    factor = 6.4
    med = col.median()
    mad = (col - med).abs().median()
    low_bound = med - mad * factor
    upp_bound = med + mad * factor
    col_cens = col[(col <= upp_bound) & (col >= low_bound)]
    extr_num = 0
    for i in col:
        if ((i > upp_bound) | (i < low_bound)):
            extr_num = extr_num + 1
    extr_num = len(col) - len(col_cens)
    return col_cens
```

Цензурирование и нормировка данных:

Цензурирование и нормировка рассматриваются на примере данных столбца "k7" (Качество кредиторской задолженности). Цензурирование выборки заключается в замене 
аномально больших и малых значений коэффициента на экономически и статистически обоснованные максимальные и минимальные пороговые значения (K_max = 0.0, K_min = 0.3).
Для возможности более удобной интерпретации результатов производится приведение значений коэффициента к шкале [0,1]. При этом более высокие значение показателя
соответствует предприятиям с более высокой степенью кредитоспособности. Так как рассматриваемый коэффициент отражает отношение просроченной кредиторской задолженности к
общей кредиторской задолженности, то для нормировки данных используется формула, соответствующая обратной зависимости кредитоспособности предприятия от рассматриваемого показателя.
Удаление аномальных наблюдений и распределение значений показателя на отрезке [0,1] отражены на новой ящичной диаграмме, соответствующей данному показателю. 

```python
def Censoring(col, k_min, k_max):
    return col.clip(k_min, k_max)

def Rationing(col, k_min, k_max):
    return (k_max - col) / (k_max - k_min)

col_excl = HampelTest(df["k7"])
col_cens = Censoring(col_excl, 0.0, 0.3)
col_rat = Rationing(col_cens, 0.0, 0.3)
df_cens = col_rat.to_frame(name="k7")
df_cens.boxplot(column="k7", grid=False, patch_artist=False)
plt.title("k7")
plt.show()
```

Вычисление парных корреляций Пирсона:

```python
def PCC(array_1, array_2):
    mean_1 = mean(array_1)
    mean_2 = mean(array_2)
    multi = array_1 * array_2
    cov = mean(multi) - mean_1 * mean_2
    return cov / (sigma(array_1) * sigma(array_2))
```

Представление результатов в виде таблицы:

```python
from decimal import Decimal
size = len(df.columns) + 1
corr_matrix = np.zeros((size, size), dtype=object)
for i in range(1, size):
    corr_matrix[0][i] = df.columns[i-1]
for i in range(1, size):
    corr_matrix[i][0] = df.columns[i-1]

for i in range(1, size):
    for j in range(1, size):
        mean_i = df[df.columns[i-1]].mean()
        mean_j = df[df.columns[j-1]].mean()
        corr_matrix[i][j] = round(PCC(df[df.columns[i-1]].fillna(mean_i), df[df.columns[j-1]].fillna(mean_j)), 2)

print (corr_matrix)
```

Представление результатов в виде DataFrame:

```python
columns = df.columns.tolist()
corr_df = pd.DataFrame(index=columns, columns=columns, dtype=object)
for i in columns:
    for j in columns:
        x = df[i].fillna(df[i].mean())
        y = df[j].fillna(df[j].mean())
        corr_df.loc[i, j] = round(PCC(x, y), 2)
print (corr_df)
```

Как видно, коэффициенты коррреляции рассчитаны верно, так в результате получаем симметричную матрицу с элементами на главной диагонали, равными 1.
При этом, все элементы матрицы не превосходят по модулю значение 1.

График рассеивания:

```python
columns = corr_df.columns.tolist()
for i in range(1, len(columns)):
    for j in range(0, i):
        if ((corr_df.iloc[i, j] >= 0.8) | (corr_df.iloc[i, j] <= -0.8)):
            df.plot.scatter(x=columns[i], y=columns[j], color='red')
            plt.title(f'График рассеивания {columns[i]} vs {columns[j]}')
            plt.xlabel(columns[i])
            plt.ylabel(columns[j])
            plt.grid(True)
            plt.tight_layout()
            plt.show()
```

Как видно из графика, наибольшая линейная зависимость наблюдается между коэффициентами k5 и k8 (К5 – коэффициент обеспеченности финансовых обязательств активами,
К8 – коэффициент финансовой независимости), которые являются коэффициентами одной группы (коэффициентами финансовой устойчивости) и определяют степень 
зависимости компании от внешних источников финансирования. При этом, наиболее целесообразно исключить из дальнейшего анализа именно показатели коэффициента к8,
поскольку данный коэффициент имеет меньшую степень важности.
